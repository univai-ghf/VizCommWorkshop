# Dont Make Me Think

There is this pretty famous book by Steve Krug, called “Dont Make Me Think”. Its a call to respect conventions for web elements, such as shopping carts (a cart should be on the upper right), so that the web experience is obvious to users.

In visualization, as in web development, your audience does not want to spend cognitive effort on things you could just show them, by convention, or by explicit writing. So, just point out the key facts and insights.

For example, in this great article in the financial times <https://www.ft.com/content/0f11b219-0f1b-420e-8188-6651d1e749ff?hcb=1>, the main point “Vaccines have made Covid-19 far less lethal” is written up-front.

![](images/pasted-image.png)

The implications are made clear in the second sentence, comparing vaccinated 80 year-olds to un-vaccinated 50 year-olds. This implication is illustrated in the visualization as well, with a horizontal black line, and a caption.

Instead of point markers, downwards pointing arrows are used on lines to reinforce the notion of lower risk. Captions and annotations are used to point out key insights. Extraneous frames and tick marks are removed.

This is an example of framing. It grabs the audience and leads it through the insights you want to share.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data2 = pd.read_csv("https://gist.github.com/rahuldave/df75d3aa73e95e8361b7cdd1663c9868/raw/721fc8d7fb48202c5e473ceb5dfc0143008e30c2/viz.csv")
data2
```

## At first, a simple but clean plot

At first we make a simple image of the data, with the red points representing the unvaxed risk, and the blue the vaxed risk. We put a rough black line to compare the danger to an unvaxed 50 year old with that of a vaxed 80 year old.

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
ax.scatter(data2.age, data2.vaxrisk, 
           alpha=0.6, s=40, marker='o', color=blue)
ax.axhline(0.037, 0, 1, color="black")
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
```

We then set up a grid like in the original plot, put labels everywhere, and get rid of external tick marks. You can read more about ticks at <https://e2eml.school/matplotlib_ticks.html> by @_brohrer_

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
ax.scatter(data2.age, data2.vaxrisk, 
           alpha=0.6, s=40, marker='o', color=blue)
ax.axhline(0.037, 0, 1, color="black")
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.tick_params(axis='y', which='minor', left=False, right=False)
```

We format the ticks the way the @FT article did, and removed the spines that make up the plot bounding box as well.

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
ax.scatter(data2.age, data2.vaxrisk, 
           alpha=0.6, s=40, marker='o', color=blue)
ax.axhline(0.037, 0, 1, color="black")
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.tick_params(axis='y', which='minor', left=False, right=False)
from matplotlib.ticker import MultipleLocator, FixedFormatter, FixedLocator
ax.xaxis.set_major_locator(MultipleLocator(base=5))
ax.yaxis.set_major_locator(FixedLocator([0.0001, 0.001, 0.01, 0.1, 1]))
ax.yaxis.set_major_formatter(FixedFormatter([0.0001, 0.001, 0.01, 0.1, "1%"]))
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
```

## Communicating the comparisons

```python
d50 = data2[data2.age==50][['age', 'risk']].values[0]
d80 = data2[data2.age==80][['age', 'vaxrisk']].values[0]
d50, d80
```

We replace the horizontal line comparing ages with an arrow, as in the original image. The arrow is drawn as a Patch, which is an aribitrary 2-D region in matplotlib. Read more about Patches in @_brohrer_'s course <https://e2eml.school/matplotlib_patches.html> and in the @matplotlib docs.

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
ax.scatter(data2.age, data2.vaxrisk, 
           alpha=0.6, s=40, marker='o', color=blue)
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.tick_params(axis='y', which='minor', left=False, right=False)
from matplotlib.ticker import MultipleLocator, FixedFormatter, FixedLocator
ax.xaxis.set_major_locator(MultipleLocator(base=5))
ax.yaxis.set_major_locator(FixedLocator([0.0001, 0.001, 0.01, 0.1, 1]))
ax.yaxis.set_major_formatter(FixedFormatter([0.0001, 0.001, 0.01, 0.1, "1%"]))
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
d50 = data2[data2.age==50][['age', 'risk']].values[0] # age, risk
d80 = data2[data2.age==80][['age', 'vaxrisk']].values[0] # age, vaxrisk
import matplotlib.patches as patches
p1 = patches.FancyArrowPatch(d50, d80, 
                             arrowstyle='<->', 
                             mutation_scale=20, 
                             linewidth=2, 
                             linestyle='--')
ax.add_patch(p1)
```

At this point, our image is beginning to take shape. But we have data points for the vaxed case instead of arrows! We'll use Patches to fix this! We also get a nice title.

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8))
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.tick_params(axis='y', which='minor', left=False, right=False)
from matplotlib.ticker import MultipleLocator, FixedFormatter, FixedLocator
ax.xaxis.set_major_locator(MultipleLocator(base=5))
ax.yaxis.set_major_locator(FixedLocator([0.0001, 0.001, 0.01, 0.1, 1]))
ax.yaxis.set_major_formatter(FixedFormatter([0.0001, 0.001, 0.01, 0.1, "1%"]))
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
d50 = data2[data2.age==50][['age', 'risk']].values[0] # age, risk
d80 = data2[data2.age==80][['age', 'vaxrisk']].values[0] # age, vaxrisk
import matplotlib.patches as patches
p1 = patches.FancyArrowPatch(d50, d80, 
                             arrowstyle='<->', 
                             mutation_scale=20, 
                             linewidth=2, 
                             linestyle='--')
ax.add_patch(p1)
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
# ax.scatter(data2.age, data2.vaxrisk, 
#            alpha=0.6, s=40, marker='o', color=blue)
for idx, s in data2.iterrows():
    p = patches.FancyArrowPatch((s.age, s.risk), (s.age, s.vaxrisk), 
                                color=blue, 
                                arrowstyle="-|>", 
                                mutation_scale=20, 
                                linewidth=2)
    ax.add_patch(p)
    ax.set_title(
"""
Vaccines have made Covid far less lethal. A fully vaccinated 80 year old
has the same risk as an unvaccinated 50 year old.
""",       
            loc="left", 
            fontsize="xx-large"
);
```

## Getting the final plot

We finally set both the figure and axes color and add 3 annotations. The annotations use arrows, which go from the xytext coordinates to the xy co-ordinates. The annotations serve as legends, and describe the main conclusion.

```python
red = "#c51d35"
blue = "#2182BF"
fig, ax = plt.subplots(figsize = (12,8), facecolor='#ffeed9')
ax.set_facecolor('#fff1e0')
ax.set_xlabel("Age", fontsize='x-large')
ax.set_xlim((25, 95))
ax.set_ylabel("Risk of catching and dying from Covid (log scale)",
             fontsize='x-large')
ax.set_ylim((0.00008, 4))
ax.set_yscale('log')
ax.grid(axis="y", color="black", alpha=.5, linewidth=.5)
ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.tick_params(labelbottom=True, labeltop=True, labelleft=True, labelright=True)
ax.tick_params(axis='y', which='minor', left=False, right=False)
from matplotlib.ticker import MultipleLocator, FixedFormatter, FixedLocator
ax.xaxis.set_major_locator(MultipleLocator(base=5))
ax.yaxis.set_major_locator(FixedLocator([0.0001, 0.001, 0.01, 0.1, 1]))
ax.yaxis.set_major_formatter(FixedFormatter([0.0001, 0.001, 0.01, 0.1, "1%"]))
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
d50 = data2[data2.age==50][['age', 'risk']].values[0] # age, risk
d80 = data2[data2.age==80][['age', 'vaxrisk']].values[0] # age, vaxrisk
import matplotlib.patches as patches
p1 = patches.FancyArrowPatch(d50, d80, 
                             arrowstyle='<->', 
                             mutation_scale=20, 
                             linewidth=2, 
                             linestyle='--')
ax.add_patch(p1)
ax.scatter(data2.age, data2.risk, 
           alpha=0.6, s=40, marker='o', color=red)
# ax.scatter(data2.age, data2.vaxrisk, 
#            alpha=0.6, s=40, marker='o', color=blue)
for idx, s in data2.iterrows():
    p = patches.FancyArrowPatch((s.age, s.risk), (s.age, s.vaxrisk), 
                                color=blue, 
                                arrowstyle="-|>", 
                                mutation_scale=20, 
                                linewidth=2)
    ax.add_patch(p)
ax.set_title(
"""Vaccines have made Covid far less lethal. A fully vaccinated 80 year old
has the same risk as an unvaccinated 50 year old.""", loc="left", fontsize="xx-large"
);
ax.annotate("Risk of Covid Death\nbefore vaccines", 
            xy=(40, data2[data2.age==40]['risk']), 
            xytext=(27, 0.04), 
            fontsize='x-large',
            color='r',
           arrowprops=dict(arrowstyle='-|>'))
ax.annotate("Risk after two doses", 
            xy=(40, data2[data2.age==40]['vaxrisk']), 
            xytext=(45, 0.0003), 
            fontsize='x-large',
            color='b',
           arrowprops=dict(arrowstyle='-|>'))
ax.annotate(
"""A fully vaccinated
80 year-old now has
roughly the same risk
of dying from covid as
an unvaccinated 50 year-old
""", 
            xy=(76, 0.046), 
            xytext=(72, 0.0003), 
            fontsize='x-large',
            color='k',
           arrowprops=dict(arrowstyle='-|>'))
fig.tight_layout()
```
